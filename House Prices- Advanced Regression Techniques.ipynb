```python
for dataset in full_data:
    # categorical features with missing values in the training set -> 
    # missing or most occuring value if "missing" has no meaning
    dataset['Alley'] = dataset['Alley'].fillna("missing")
    dataset['BsmtCond'] = dataset['BsmtCond'].fillna("missing")
    dataset['BsmtExposure'] = dataset['BsmtExposure'].fillna("missing")
    dataset['BsmtFinType1'] = dataset['BsmtFinType1'].fillna("missing")
    dataset['BsmtFinType2'] = dataset['BsmtFinType2'].fillna("missing")
    dataset['BsmtQual'] = dataset['BsmtQual'].fillna("missing")
    dataset['Electrical'] = dataset['Electrical'].fillna("SBrkr")
    dataset['Fence'] = dataset['Fence'].fillna("missing")
    dataset['FireplaceQu'] = dataset['FireplaceQu'].fillna("missing")
    dataset['GarageCond'] = dataset['GarageCond'].fillna("missing")
    dataset['GarageFinish'] = dataset['GarageFinish'].fillna("missing")
    dataset['GarageQual'] = dataset['GarageQual'].fillna("missing")
    dataset['GarageType'] = dataset['GarageType'].fillna("missing")
    dataset['MasVnrType'] = dataset['MasVnrType'].fillna("None")
    dataset['MiscFeature'] = dataset['MiscFeature'].fillna("missing")
    dataset['PoolQC'] = dataset['PoolQC'].fillna("missing")
    # categorical features with no missing values in the training set -> most occurring value
    cat_2_columns = ['BldgType', 'CentralAir', 'Condition1', 'Condition2', 'ExterCond', 'ExterQual',
                     'Exterior1st', 'Exterior2nd', 'Foundation', 'Functional', 'Heating', 'HeatingQC', 
                     'HouseStyle', 'KitchenQual', 'LandContour', 'LandSlope', 'LotConfig', 'LotShape', 
                     'MSZoning', 'MasVnrType', 'Neighborhood', 'PavedDrive', 'RoofMatl', 'RoofStyle',
                     'SaleCondition', 'SaleType', 'Street', 'Utilities']
    for col in cat_2_columns:
        dataset[col] = dataset[col].fillna(train[col].mode()[0])
    
```
