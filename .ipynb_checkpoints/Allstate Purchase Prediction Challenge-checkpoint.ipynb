```python
profile = pandas_profiling.ProfileReport(train_data)
profile
```

```python
def get_hour(row_col):
    return row_col.hour

train_data.drop(profile.get_rejected_variables(), inplace=True)
train_data.drop(['customer_ID'], inplace=True)
train_data['time'] = train_data.time.apply(pd.to_datetime, format='%H:%M')
train_data['hour'] = train_data.time.apply(get_hour)
train_data.drop(['time'], inplace=True)

```
